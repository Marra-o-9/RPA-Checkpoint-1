```python
import pandas as pd
import sqlite3
```

```python
# Função para carregar dados do Excel e inserir no SQLite
def excel_to_sqlite(excel_file, db_name, table_name):
    # Carregar o arquivo Excel para um DataFrame do pandas
    df = pd.read_excel(excel_file)

    # Conectar ao banco de dados SQLite (ou criar se não existir)
    conn = sqlite3.connect(db_name)
    
    # Criar uma tabela no SQLite com base nas colunas do DataFrame
    df.to_sql(table_name, conn, if_exists='replace', index=False)

    # Confirmar as alterações e fechar a conexão
    conn.commit()
    conn.close()
    print(f"Dados do arquivo {excel_file} foram inseridos na tabela {table_name} do banco {db_name}.")
```

```python
# Função para carregar e exibir dados do banco SQLite
def display_data_from_sqlite(db_name, table_name):
    # Conectar ao banco de dados SQLite
    conn = sqlite3.connect(db_name)
    
    # Ler os dados da tabela do SQLite
    query = f"SELECT * FROM {table_name}"
    df = pd.read_sql_query(query, conn)

    # Fechar a conexão com o banco de dados
    conn.close()

    # Exibir o DataFrame no Jupyter Notebook
    return df
```

```python
# Especifique o caminho do arquivo Excel, banco de dados e nome da tabela
excel_file = 'sprint_AU\\vagas_infojobs.xlsx'  # Substitua pelo caminho do seu arquivo
db_name = 'vagas_infojobs.db'          # Nome do banco de dados SQLite
table_name = 'vagas'       # Nome da tabela que será criada no SQLite

# Chamar a função para realizar a conversão
excel_to_sqlite(excel_file, db_name, table_name)
```

```python
# Exibir os dados no Jupyter Notebook
df = display_data_from_sqlite(db_name, table_name)
df  # O DataFrame será automaticamente renderizado como tabela no Jupyter Notebook
```
